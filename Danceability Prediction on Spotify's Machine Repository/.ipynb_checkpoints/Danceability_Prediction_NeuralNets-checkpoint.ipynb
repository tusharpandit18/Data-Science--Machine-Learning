- Read the data and print a few rows

```python
import pandas as pd

data = pd.read_csv('Danceability_Prediction_Spotify.csv', skipinitialspace=True)
```

- get a sense of the size of the dataset by printing the number of rows and columns

```python
rows,columns = data.shape
print('Number of rows:', rows)
print('Number of columns:', columns)
```

# Data Preprocessing

- Convert all data to numeric and replace string values by NaN

```python
for c in data.columns:
        data[c] = pd.to_numeric(data[c], errors='coerce')
```

- normalize the data (convert all column values in the range of 0 to 1) so that all the values are on the same scale

```python
df_norm = data.sub(data.min()).div((data.max() - data.min()))
data = df_norm
data.dance = data.dance.astype(int)
```

```python
data.median()
```

# Determining correlations

- Plot scatter plot

```python
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sb
sb.set(style='whitegrid', context='notebook')
columns = ['energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','danceability']
sb.pairplot(data[columns])
plot.show()
```

- Plot heatmap so that we can actually compare the correlation values

```python
import numpy as np
cor_matrix = np.corrcoef(data[columns].values.T)
sb.set(font_scale=1.5)
cor_heat_map = sb.heatmap(cor_matrix, cbar=True, annot=True,square=True,fmt='.2f', annot_kws={'size':9},
yticklabels=columns,
xticklabels=columns)
plot.show()
data.drop('danceability', axis=1, inplace=True)
data.rename(columns={'dance': 'danceability'}, inplace=True)
```

- Check the correlation by computing the mean just to verify the relationships

```python
data.groupby('danceability').mean()
```

```python
data.shape
```

- Delete columns that have all NaN's (eliminate string columns)

```python
data = data.loc[:, data.isnull().mean() < 1]
```

- Segregate feature vector and class labels

```python
data.dropna()
# X is the feature vector
X = data[['valence','energy','loudness','acousticness']].values
# y is the class label
y = data['danceability'].values
```

- Print feature vector X and class label y

```python
print X
print y
```

- 5-fold cross validation and train-test split

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import decomposition

CV = 5
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
```

- Fit model without PCA and compute training and testing accuracy

```python
mlp = MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
scores = cross_val_score(mlp, X_train, y_train, cv=CV)
print("Training Accuracy without PCA : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn import metrics
predicted = cross_val_predict(mlp, X_test, y_test, cv=CV)
print("Testing Accuracy without PCA : %0.2f" %metrics.accuracy_score(y_test, predicted))
```

- Fit model with PCA and compute training and testing accuracy

```python
X = data.iloc[:, :-2].values
y = data.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)


pca = decomposition.PCA(n_components=4)
pca.fit(X_train)
X_train = pca.transform(X_train)

mlp = MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
scores = cross_val_score(mlp, X_train, y_train, cv=CV)
print("Training Accuracy with PCA: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn import metrics
predicted = cross_val_predict(mlp, X_test, y_test, cv=CV)
print("Testing Accuracy with PCA : %0.2f" %metrics.accuracy_score(y_test, predicted))
```

- Classification Report

```python
from sklearn.metrics import classification_report
from sklearn.cross_validation import cross_val_predict
y_pred = cross_val_predict(mlp, X, y, cv=10)
print(classification_report(y, y_pred))

```
